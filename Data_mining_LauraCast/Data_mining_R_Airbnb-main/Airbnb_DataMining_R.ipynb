El barrio parece ser un indicador importante para los metros cuadrados de un apartamento.
Agrupamos los barrios por metros cuadrados. Usamos una matriz de similaridad de Tukey:

```r
tky<-TukeyHSD(aov( formula=Square.Meters~Neighbourhood, data=df_madrid ))
tky.result<-data.frame(tky$Neighbourhood)
cn <-sort(unique(df_madrid$Neighbourhood))
resm <- matrix(NA, length(cn),length(cn))
rownames(resm) <- cn
colnames(resm) <- cn
resm[lower.tri(resm) ] <- round(tky.result$p.adj,4)
resm[upper.tri(resm) ] <- t(resm)[upper.tri(resm)] 
diag(resm) <- 1
library(ggplot2)
library(reshape2)
dfResm <- melt(resm)
ggplot(dfResm, aes(x=Var1, y=Var2, fill=value))+
  geom_tile(colour = "black")+
  scale_fill_gradient(low = "white",high = "steelblue")+
  ylab("Class")+xlab("Class")+theme_bw()+
  theme(axis.text.x = element_text(angle = 90, hjust = 1),legend.position="none")
```

Dibujamos un dendrograma de los diferentes barrios:


```r
df_madrid.dist <- as.dist(1 - abs(resm)) #distancia, resm de la matriz Tukey
df_madrid.tree <- hclust(df_madrid.dist, method="complete")
df_madrid.dend <- as.dendrogram(df_madrid.tree)
```

```r
library(dendextend)
plot(color_branches(df_madrid.dend, h=0.2))
abline(h=0.2,col="red")
```

Elegimos el punto de corte aconsejable y mostramos los 3 clusters:

```r
clusters<- cutree(df_madrid.dend, k=3)
print("Aparecen 3 clusters, que son: ")

c1 <-names(clusters[clusters==1])
cat('\ncluster1 -> \n', c1)
c2 <-names(clusters[clusters==2])
cat('\ncluster2 ->\n', c2)
c3 <-names(clusters[clusters==3])
cat('\ncluster3 ->\n', c3)

print('El punto de corte ideal será en ')
```

```r
library(cluster)

ss<-silhouette(clusters, df_madrid.dist)
plot(ss,col=1:max(clusters),border=NA)
```

creamos una nueva columna en el dataframe df_madrid ('neighb_id') con un nuevo identificador marcado por los clusters obtenidos:

```r
neighb_id  <- as.factor(cutree(df_madrid.dend, k=3))
df_madrid_cl <- data.frame(names(neighb_id), neighb_id)
colnames(df_madrid_cl) <- c('Neighbourhood','neighb_id')
df_madrid_cl
```

Vamos a crear dos grupos, uno test y otro train.

Tratamos de predecir los metros cuadrados en función del resto de variables del dataframe.

```r
#elimino los NA de Square.Meters
library(tidyverse)
df_mad<-df_madrid %>% drop_na(Square.Meters)
```

```r
#dividir los datos de df_madrid en dos grupos
set.seed(1)

itrain<-sample(1:nrow(df_mad),round(nrow(df_mad)*0.7))
madrid.train<- df_mad[itrain,]
madrid.test <- df_mad[-itrain,]
```

```r
#creamos el modelo
#excluimos las variables que no influyen en los metros cuadrados

model<-lm(Square.Meters~
           Bathrooms+Bedrooms+Price+Square.Meters+neighb_id, data = madrid.train)
summary(model)

#beta
#beta<-model$coefficients
#beta
```

```r
confint(model)
```

```r
#creamos la columna con los valores del predict
madrid.train$pred<-predict(model,madrid.train)
caret::postResample(pred=madrid.train$pred, obs=madrid.train$Square.Meters)

plot(madrid.train$Square.Meters, (madrid.train$Square.Meters-madrid.train$pred), xlab = "Square.Meters", ylab = "Residuos")
caret::postResample(pred=madrid.train$pred, obs= madrid.train$Square.Meters)
```

Observamos el histograma de los residuos sobre el conjunto de test para evaluar la calidad de vuestro modelo

```r
#probarlo con los datos de testing
madrid.test$pred<-predict(model,madrid.test)
plot(madrid.test$Square.Meters, (madrid.test$Square.Meters-madrid.test$pred))

caret::postResample(pred=madrid.test$pred, obs=madrid.test$Square.Meters)
```

```r
#representar el histograma de residuos
hist(madrid.test$Square.Meters-madrid.test$pred, 30)
qqnorm(madrid.test$Square.Meters-madrid.test$pred)
qqline(madrid.test$Square.Meters-madrid.test$pred, col = 'orange', lwd =2)
```

```r
#probar la distancia de cooks
plot(cooks.distance(model), 
     xaxp = c(0, 120, 14),
     yaxp = c(0, 0.8, 8))
```

```r
madrid.train_clean<-madrid.train[cooks.distance(model)<0.05,]
model_clean<-lm(Square.Meters~
           Bathrooms+Bedrooms+Price+Square.Meters+neighb_id,data=madrid.train_clean)
summary(model_clean)
```

```r
myrsquared<-function(Y,est_Y){
    Rsq <- 1-(sum((Y-est_Y)^2))/(sum((Y-mean(Y))^2))
    return(Rsq)
}
```

```r
madrid.test$pred<-predict(model_clean,madrid.test)

plot(madrid.test$Square.Meters,(madrid.test$Square.Meters-madrid.test$pred))
hist(madrid.test$Square.Meters-madrid.test$pred,30)
paste("MSE:", sqrt(mean((madrid.test$Square.Meters-madrid.test$pred)^2)))
paste("R^2:", myrsquared(madrid.test$Square.Meters,madrid.test$pred))


```

Nos preguntamos si tuvieramos un apartamento para 6 personas, con 1 baño, con un precio de 80€/noche y 3 habitaciones en el barrio de Sol, con 3 camas y un review de 80.
¿Cuantos metros cuadrados tendría?¿Como varía sus metros cuadrados con cada habitación adicional?

```r
madrid_pred <- subset(df_madrid, select = c('Neighbourhood','Accommodates','Bathrooms','Bedrooms','Beds','Price','Review.Scores.Rating', 'Square.Meters'))


madrid_pred$pred_SqM <- predict(model_clean, df_madrid)

madrid_pred$residual <- madrid_pred$Square.Meters - madrid_pred$pred_SqM
head(madrid_pred)
```

```r
#impiar datos de valores nulos para poder hacer cv
limpio <- na.omit(madrid_pred)

library(glmnet)

#hago la cv
cv_mad_pred<-cv.glmnet(data.matrix(limpio[,c('Accommodates','Bathrooms','Bedrooms','Beds','Price','Review.Scores.Rating', 'residual')]),data.matrix(limpio[,'Square.Meters']),nfolds=10,alpha=1, standarize=T)
plot(cv_mad_pred)
```

```r
meters <- predict(cv_mad_pred, newx=as.matrix(data.frame(Accommodates=6, Bathrooms=1, Bedrooms=3,  Beds= 3, Price=80, Review= 80, sv=cv_mad_pred$lambda.1se)))
paste("Los metros cuadrados del piso serían: ", meters,'metros')

meters2 <-predict(cv_mad_pred, newx=as.matrix(data.frame(Accommodates=6, Bathrooms=1, Bedrooms=4,  Beds= 3, Price=80, Review= 80, sv=cv_mad_pred$lambda.1se)))
paste("Los metros cuadrados del piso2 serían: ", meters2,'metros')

paste("La diferencia de metros entre los dos con solo una habitación más es:", meters - meters2,'metros')

```

```r
#Tengo la columna pred_SqM ya hecha!!! podemos sustituirla para todos los valores NA de Square.Meters

#todas las filas de Square.Meters con valores
df_sqm_valores <- filter(df_madrid, !is.na(df_madrid$Square.Meters))

#extraer del df los valores a sustiuir 
sustituir_sqm_pred_madrid <- filter(df_madrid, is.na(df_madrid$Square.Meters))


#sustituir los NA por el valor predicho
sustituir_sqm_pred_madrid$Square.Meters <- predict(model_clean, sustituir_sqm_pred_madrid)


#combinar los sustituidos con los valores reales
df_madrid_sqm_sustuidos<-rbind(sustituir_sqm_pred_madrid,df_sqm_valores)

#acorto el nombre
df_mad_sqm <- df_madrid_sqm_sustuidos
df_mad_sqm
```

Utilizamos PCA para encontrar el apartamento más cercano a uno dado.

Este algoritmo nos ayudaría a dado un apartamento, el algoritmo nos devolvería los 5 apartamentos más similares.

```r
#limpiamos de NAs el df con los valores predichos de SqM

df_mad_sqm_clean  <-df_mad_sqm[, c(
    'Accommodates',
    'Bathrooms',
    'Bedrooms',
    'Beds',
    'Price',
    'Guests.Included',
    'Extra.People',
    'Review.Scores.Rating',
    'Latitude',
    'Longitude',
    'Square.Meters'
)]

df_mad_pca<- na.omit(df_mad_sqm_clean)
head(df_mad_pca)
```

```r
##Utilizamos prcomp para calcular el pca

pmadrid<-prcomp(df_mad_pca,center = TRUE, scale = TRUE)
pmadrid
```

```r
#pintar los autovalores 

plot(cumsum(pmadrid$sdev^2/sum(pmadrid$sdev^2)),main="Autovalores")
```

```r
#Matrix de componentes principales
dim(pmadrid$x)
```

```r
#nos quedamos con 4 componentes
number_of_pca_components <-4
Apc<-pmadrid$x[,1:number_of_pca_components]
dist<-rep(NA,nrow(Apc))
head(Apc)
```

```r
piso <- data.frame(Accommodates=6 , Bathrooms=1.0, Bedrooms=2, Beds=3, Price=60, Guests.Included=2, Extra.People=10, Review.Scores.Rating=88 , Latitude=40.40554, Longitude=-3.710429, Square.Meters= 67.43726)


predict<-matrix(predict(pmadrid, piso)[1:number_of_pca_components],nrow=1)


for (i in 1:nrow(Apc)){
    dist[i]<-sum((predict-Apc[i,])^2)
}


top_parecidos <- df_mad_pca[head(order(dist,decreasing = FALSE),6), ]
paste('El piso escogido era: ')
top_parecidos[1, ]
paste(' Los 5 más parecidos son: ')
top_parecidos[2:6, ]
    
```
