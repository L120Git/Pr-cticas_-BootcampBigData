Cual es el precio medio de alquiler de cada uno, la diferencia que hay ¿es estadísticamente significativa?
¿Con que test lo comprobarías?

```r
#sacar los precios de cada tipo de alquiler
#Entire room/apt
preciosEntire <- filter(airbnbClean, TipoAlquiler == 'Entire home/apt')
PVPEntire <- preciosEntire$Precio
PVPEntire=na.omit(PVPEntire)

#Private room
preciosPrivate <- filter(airbnbClean, TipoAlquiler == 'Private room')
PVPPrivate <- preciosPrivate$Precio
PVPPrivate=na.omit(PVPPrivate)

#Shared room
preciosShared<- filter(airbnbClean, TipoAlquiler == 'Shared room')
PVPShared <- preciosShared$Precio
#Limpiar preciosShared
PVPShared=na.omit(PVPShared)

#calcular la media de cada conjuto de precios 
media<-function(X){
    m <- 0
    for (xi in X){
        m <- m + xi
    }
    m/length(X)
}

paste("La media de precios de Entire room/apt es: ", round(media(PVPEntire),2), "€")
paste("La media de precios de Private room es: ", round(media(PVPPrivate),2), "€")
paste("La media de precios de Shared room es: ", round(media(PVPShared),2), "€")

#saber si siguen una distribución normal
qqnorm(PVPEntire)
qqline(PVPEntire, col = 'green', lwd =2)
qqnorm(PVPPrivate)
qqline(PVPPrivate, col = 'green', lwd =2)
qqnorm(PVPShared)
qqline(PVPShared, col = 'green', lwd =2)


#test kruskal-wallis porque no siguen una distribución normal

kruskal.test( Precio ~ TipoAlquiler, data=airbnbClean)

print("El pvalor es: < 2.2e-16. La diferencia parece que es estadisticaente significativa")
```

Ayuda: Usa la función aggregate `aggregate(.~colname,df,mean,na.rm=TRUE)`

```r
#calcular el precio medio de airbnb_entire$Precio
PreciosEnt<- airbnb_entire$Precio
Precio_mean <- aggregate(PreciosEnt ~ (Barrio), data = airbnb_entire, mean,na.rm=TRUE)


#ordenar en descreciente
Precio_mean_Desc <- Precio_mean[with(Precio_mean, order(-PreciosEnt)), ]

#sacar los 5 primeros

primerosBarriosPrecio_mean <- Precio_mean_Desc[1:5,]
primerosBarriosPrecio_mean
```

Calcular un modelo lineal que combine alguna de estas variables:
* NumBanyos
* NumDormitorios
* MaxOcupantes
* MetrosCuadrados

```r
barrio_sol<-subset(airbnb_entire,Barrio=="Sol")

#Con los datos en bruto:

#Relación entre MetrosCuadrados y MaxOcupantes
library(ggplot2)
ggplot(barrio_sol, aes(x=MetrosCuadrados, y=MaxOcupantes)) + 
  geom_point() + 
    geom_smooth(method='lm', formula=y~x, se=FALSE, col='dodgerblue1') + 
        theme_light()

#Limpiar los outayers eliminado 0m2 & m2>150
barrio_solL <- barrio_sol %>% filter(MetrosCuadrados>0 & NumDormitorios>0)

#Modelo lineal
lmSol <- lm(MaxOcupantes ~ MetrosCuadrados+NumDormitorios+NumBanyos+MaxOcupantes, data=barrio_solL)

#todos los datos del lm
summary(lmSol)
confint(lmSol)
```

Primero calculamos la correlación para ver como se relacionan estas variables entre sí.

```r
#Limpiar m2 
m2<-factor(barrio_sol$MetrosCuadrados)
m2[is.na(m2)] <- 0
m2 <- as.numeric(m2)

#Limpiar maxCup
maxOcup <-barrio_sol$MaxOcupantes
maxOcup= na.omit(maxOcup)
maxOcup <- as.numeric(maxOcup)

#Limpiar habs
habs <-barrio_sol$NumDormitorios
habs= na.omit(habs)
habs <- as.numeric(habs)

#Limpiar baños
baños <-barrio_sol$NumBanyos
baños= na.omit(baños)
baños <- as.numeric(baños)

#Calcular la correlación
print(paste("La correlación de m2 y MaxOcupantes, es:",cor(m2,maxOcup)))
print(paste("La correlación de m2 y habs, es:",cor(m2,habs)))
print(paste("La correlación de m2 y baños, es:",cor(m2,baños)))
```
